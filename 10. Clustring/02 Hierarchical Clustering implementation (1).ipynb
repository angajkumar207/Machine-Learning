## Hierarchical Clustering implementation



```python
# Import required labraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
data = datasets.load_iris()
x = data.data
y = data.target
```

```python
x
```

```python
y  # This is unsupervised lerning algorithm problem there for y will not used
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler
```

```python
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
## Applying the pca

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_scaled = pca.fit_transform(x_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = y)
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1])
```

```python
# Agglomerative clustring >> make dendogram

import scipy.cluster.hierarchy as  sc

plt.figure(figsize=(12, 6))
plt.title("Dendogram")

sc.dendrogram(sc.linkage(pca_scaled, method="ward"))

plt.xlabel("sample data")
plt.ylabel("Euclidean distance")

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, metric="euclidean", linkage="ward")
cluster
```

```python
# fit
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:, 0], pca_scaled[:, 1], c = cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score

silhouette_coef = []
for k in range(2, 11):
    agglo = AgglomerativeClustering(n_clusters=k, metric='euclidean', linkage='ward')
    agglo.fit(x_scaled)
    score = silhouette_score(x_scaled, agglo.labels_)
    silhouette_coef.append(score)
```
