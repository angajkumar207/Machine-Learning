```python
## Import Library
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"healthcare.csv")
```

```python
df
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
out = IsolationForest(contamination=0.2)
out
```

```python
# fit
out.fit(df)
```

```python
pred = out.predict(df)
```

```python
pred
```

```python
index = np.where(pred<0)
index
```

```python
df
```

```python
x = df.values
x
```

```python
plt.scatter(df.iloc[:,0], df.iloc[:,1])
```

```python
plt.scatter(x[index, 0], x[index, 1], edgecolors='red')
```
