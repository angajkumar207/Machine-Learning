### Dbscan for outliers

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles # non linear data
```

```python
x, y = make_circles(n_samples=800, factor=0.3, noise=1)
```

```python
x
```

```python
plt.scatter(x[:,0], x[:,1])
```

```python
dbscan = DBSCAN(eps=0.1)
dbscan.fit_predict(x)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:, 0], x[:,1], c = dbscan.labels_)
```
