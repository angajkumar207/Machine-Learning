```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import make_classification
x,y = make_classification(n_samples = 1000, n_features=2, n_redundant=0, n_classes=2, random_state = 1)
```

```python
x.shape
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.30, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier()
clf
```

```python
## fit the model
clf.fit(x_train, y_train)
```

```python
## predict the data 
y_pred= clf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

## print
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Hyperprameter tunning

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_neighbors' :  [3, 5, 6, 7, 8, 9, 11, 13],
     'weights' : ['unifrom', "distance"],
     'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'],
     'leaf_size' : [20, 30 ,40, 50]
}
```

```python
grid = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, verbose=3)
grid
```

## Tain the model 

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
## predict with the best estimator
best_model = grid.best_estimator_
y_pred_tuned =best_model.predict(x_test)
```

```python
y_pred_tuned
```

```python
## Evaluate the tuned model
print("Tuned model performance")
print(f"Accuracy: {accuracy_score(y_test, y_pred_tuned)}")
print("classification report")
print(f"Classification report : {classification_report(y_test, y_pred_tuned)}")
print("confusion matrix")
print(f"Confusion matrix: {confusion_matrix(y_test, y_pred_tuned)}")
```
