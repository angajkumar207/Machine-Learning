```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
from  sklearn.datasets import make_blobs 

import warnings
warnings.filterwarnings("ignore")
```

### Genrate the dataset

```python
x, _  = make_blobs(n_samples=500, centers=3, cluster_std=0.65, random_state=44)

plt.figure(figsize=(12, 6))
plt.scatter(x[:, 0], x[:, 1], s=50)
plt.title("Unlabeled Synthetic Data")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

### Kmeans clustring

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=44)
label_kmeans = kmeans.fit_predict(x)
```

### Create the elbow graph

```python
WCSS = []

k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=44)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)

plt.figure(figsize=(12, 6))
plt.plot(k_range, WCSS, marker="o")
plt.title("Elbow Methods for optimal K")
plt.xlabel("NUMBER OF CLUSTER (K)")
plt.ylabel("Within cluster Sum of square (WCSS) ")
plt.show()
```

### Hierachial Clustring

```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

hierachial_model = AgglomerativeClustering(n_clusters=3)
label_hierachial = hierachial_model.fit_predict(x)

plt.scatter(x[:,0], x[:,1], c=label_hierachial, s=50)
plt.title("Hierachial Cluster Result")

plt.show()
```

### DBSCAN

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
label_dbscan = dbscan.fit_predict(x)

plt.scatter(x[:,0], x[:,1], c=label_dbscan, s=50)
plt.title("Dbscan Clustring Algorithem")
plt.show()
```
